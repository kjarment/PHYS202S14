
(a) Use the numerical solution to see what happens to the period of the pendulum as you change the initial angle of release. Plot θ vs. t for 5 different initial angles, θ0 , spanning 0 to π. (i.e. 0, π/4 , π/2 , 3π/4 , and π ). Put them all on the same plot. What happens to the period of the pendulum as the initial angle increases from π/4 to 3π/4? What happens to the trajectory when the initial angle is 0 or π ?

```python
def deriv(theta_current,t):  # return derivatives of the array theta
    g = 9.8 #m/s^2
    ell = 0.5 #m
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta = omega
    domega = -(g/ell)*np.sin(theta)
    return np.array([ dtheta, domega ])
```

```python
def Pendulum(theta, color = "r"):
    g = 9.8
    ell = 0.5
    t = np.arange(0.,10. + 0.01,0.01)
    theta_init = np.array([theta, 0]) #inital values
    trutheta = odeint(deriv,theta_init,t)
    plt.plot(t, trutheta, color)

Pendulum(0, 'r')
Pendulum(np.pi/4,'b')
Pendulum(np.pi/2, 'm')
Pendulum(3*np.pi/4, 'g')
Pendulum(np.pi, 'y')

plt.show()
```

```python
def EulerMeth(dx):
    xmax = 1+dx
    nsteps = int(xmax/dx)
    x = np.zeros(nsteps)
    y = np.zeros(nsteps)
    
    x[0] = 0.
    y[0] = 0.
    
    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        y[i+1] = 2*y[i] + x[i]
    
    return x, y

xe, ye = EulerMeth(.25)
xd, yd = EulerMeth(.02)
print xe, ye    
```

(a) Create a derivative function to solve this system of equations and use odeint to solve it for a projectile fired at an angle of 50 degrees from the horizontal at an initial speed of 30 m/s. Track the projectile's motion over the time interval of 0 to 200 seconds and let (x0,y0) = (0,0). You'll have to calculate (vx0,vy0) from the information provided. Let vter=44 m/s. Plot the trajectory of the projectile (y vs. x) to see how far it can travel in the x direction before landing.

```python
t3 = np.arange(0.,200.+0.1,0.1)
v0x = 30 * np.cos(50 * (np.pi/180))
v0y = 30 * np.sin(50 * (np.pi/180))
x0 = 0
y0 = 0
loc = [x0, y0, v0x, v0y]
vter = 44.
g= 9.8

def derivproj(loc, t):
    
    xveloc = loc[2]
    yveloc = loc[3]
    xaccel = -g*loc[2]/(vter**2)*np.sqrt(loc[2]**2 + loc[3]**2)
    yaccel = -g*(1 + (loc[3]/(vter**2)*np.sqrt(loc[2]**2) + loc[3]**2))
    
    array = [xveloc, yveloc, xaccel, yaccel]
    
    return array

len(t3)
```

```python
integrals = odeint(derivproj, loc, t)
```

```python
xpos = integrals[:,0]
ypos = integrals[:,1]
xvel = integrals[:,2]
yvel = integrals[:,3]

plt.plot(xpos, ypos)
plt.plot(xvel,yvel)
plt.show()
```
