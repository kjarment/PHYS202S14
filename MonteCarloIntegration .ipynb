```python
def hitOrMiss(N=1000,xmin=1.7,xmax=12.4):
    
    #Set up variables and functions
    f = lambda x: x**(0.5) + np.sin(x)
    x = np.arange(0.,15.,0.001)
    #integral of f
    Iana = lambda x: (2./3.)*x**(3./2.) - np.cos(x)
    #Numerical answer to integral
    Iexact = Iana(xmax) - Iana(xmin)
    
    #Find the max value of f within the boundary to set ymax
    #subx is a boolean mask
    subx = x[logical_and((x>xmin),(x<xmax))]
    ymax = f(subx).max()*1.05
    ymin = 0.
    
    #########################################################
    #INTEGRATION PART
    
    samples_x = xmin + (xmax-xmin)*np.random.random_sample(N);
    samples_y = ymin + (ymax-ymin)*np.random.random_sample(N);
    
    newmask = (samples_y < f(samples_x))
    I = (xmax-xmin)*(ymax-ymin)*newmask.sum()/newmask.size
    Ierr = 100*np.abs(Iexact-I)/Iexact
    #integrating with a boolean mask
    #########################################################
    
    #The results of the integration
    print "I_exact =",Iexact
    print "I_hitormiss =",I
    print "Error on I =",Ierr,"%"
    
    #plotting the function 
    plt.plot(x,f(x), lw=3)
    plt.ylabel(r'$x^{1/2} + \sin(x)$',fontsize=20)
    plt.xlabel(r'$x$', fontsize=20)
    plt.xlim(0.,15.)
    
    
    #Just showing first 500 points
    plt.plot(samples_x[1:500],samples_y[1:500],'g.')
    
    plt.plot(samples_x[newmask[:500]], samples_y[newmask[:500]],'r.')
    
    #plot the box
    plt.plot([xmin,xmin], [ymin,ymax], color='k', linestyle='--')
    plt.plot([xmax,xmax], [ymin,ymax], color='k', linestyle='--')
    plt.plot([xmin,xmax], [ymax,ymax], color='k', linestyle='--')
    
    plt.show()
    
V = interact(hitOrMiss,N=(1000,100000, 100), xmin=(0.,15.,0.001),
             xmax=(0.,15.,0.001))
    
    
```

```python
def sampleMean(N=10000,xmin=1.7,xmax=12.4):
    
    #Set up functions and variables
    f = lambda x: x**(0.5) + np.sin(x)
    x = np.arange(0.,15.,0.001)
    Iana = lambda x: (2./3.)*x**(3./2.) - np.cos(x)
    Iexact = Iana(xmax)-Iana(xmin)
    
    ###########################################################
    #INTEGRATION PART
    samples_x = xmin + (xmax-xmin)*np.random.random_sample(N);
    
    approx = f(samples_x)
    
    ###This is the sample mean method!!! THIS IS THE ACTUAL INTEGRAL
    I = approx.mean()*(xmax-xmin)
    
    #Plug in numbers into uncertainty equation
    Iunc = (xmax-xmin)*np.sqrt( ((approx**2).mean()-(approx.mean())
                                 **2)/N )
    
    #The error
    Ierr = 100*np.abs(Iexact-I)/Iexact
    ############################################################
    
    
    #find max value of f w/in boundary
    
    ymax = approx.max()
    ymin = 0.
    
    #plot the results
    print "I_exact =", Iexact
    print "I_samplemean =", I, "+/- ",Iunc
    print "Error on I =", Ierr, "%"
    
    #plot the function
    plt.plot(x,f(x),lw=3)
    plt.ylabel(r'$x^{1/2} + \sin(x)$',fontsize=20)
    plt.xlabel(r'$x$',fontsize=20)
    plt.xlim(0.,15.)
    
    plt.plot([xmin,xmin], [ymin,ymax], color='k', linestyle='--')
    plt.plot([xmax,xmax], [ymin,ymax], color='k', linestyle='--')
    plt.plot([xmin,xmax], [ymax,ymax], color='k', linestyle='--')
    
    #Average value
    plt.plot((xmin,xmax), (approx.mean(),approx.mean()), 'r-',
             label = "Average value of function at sample points")
    plt.legend(loc="upper left")
    
    plt.show()
    
v2 = interact(sampleMean, N=(1000,100000,100),xmin=(0.,15.,0.001)
              ,xmax=(0.,15.,0.001))
```

```python
fxyz = lambda x,y,z: 4. - x**2 - y**2 - z**2
xa = 0.; xb = 9./10.
ya = 0.; yb = 1.
za = 0.; zb = 11./10.

i = (10,100,1000,10000,100000,1000000)
for N in i:
    x = (xb-xa)*np.random.random_sample(N)
    y = (yb-ya)*np.random.random_sample(N)
    z = (zb-za)*np.random.random_sample(N)
    
    I = (xb-xa)*(yb-ya)*(zb-za)*fxyz(x,y,z).mean()
    err = (xb-xa)*(yb-ya)*(zb-za)*np.sqrt(((fxyz(x,y,z)**2).mean()
                                           -(fxyz(x,y,z).mean())**2
                                           )/N)
    exact = 14817./5000
    
    print "Exact =",exact,"I=",I," +/-",err,"percentErr =",100.*np.abs(I-exact)/exact
    
```

```python
func5D = lambda x1,x2,x3,x4,x5: 6 - x1**2 - x2**2 - x3**2 - x4**2 - x5**2

L = 7./10; W = 4./5; H = 9./10; P = 1.; Q = 11./10

exact = 63987./25000

for N in (10,100,1000,10000,100000,1000000):
    x1 = L*np.random.random_sample(N)
    x2 = W*np.random.random_sample(N)
    x3 = H*np.random.random_sample(N)
    x4 = P*np.random.random_sample(N)
    x5 = Q*np.random.random_sample(N)
    
    I = (L*W*H*P*Q)*func5D(x1,x2,x3,x4,x5).mean()
    #uncertainty formula
    unc = L*W*H*P*Q*np.sqrt((func5D(x1,x2,x3,x4,x5)**2).mean() - 
                            (func5D(x1,x2,x3,x4,x5).mean())**2)
    err = 100*np.abs(I-exact)/exact
    
    print "Integral=%.5f, +/- %.6f, percent err = %.3f, N=%d" %(I,unc,err,N)
```
